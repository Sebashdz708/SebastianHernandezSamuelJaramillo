```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importar los datos
datos = np.loadtxt('EnergiaPotencialGas2D.txt')
tiempo = datos[:, 0]
energia = datos[:, 1]
valor_medio = np.mean(energia)
energia_mod = energia - valor_medio
```

```python
# Realizar la transformada de Fourier
transform = np.fft.fft(energia_mod)
fundamental = np.argmax(transfor)

tiempo_medio = 1/(fundamental/len(tiempo))

mask = transform != transform[fundamental]

transform[mask] = 0

inversa = np.fft.ifft(transform)
inversa = np.real(inversa)
```

```python
# Graficaci√≥n
plt.figure()
#plt.plot(tiempo, energia, linestyle='dashed')
plt.plot(tiempo, energia_mod, linestyle ='dashed', label ='Serie Original')
plt.plot(tiempo, inversa, linestyle = 'solid', c= 'darkorange', label='Frecuencia Fundamental')
plt.title('Tiempo libre {:.2f} pasos temporales'.format(tiempo_medio))
plt.xlabel('$t [s]$')
plt.ylabel('$E[J]$')
plt.legend()
```
